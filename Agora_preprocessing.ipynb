```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from pandas import DataFrame
import os

import re

import time
from dateutil.relativedelta import relativedelta
from datetime import datetime
```

```python
path = "D:/agora.tar/agora/"
```

# Product 

```python
p_file_name_list = []
product_name_list = []
vendor_name_list = []
img_src_list = []
value_list = []
from_list = []
to_list = []
context_list = []
category1_list = []
category2_list = []
category3_list = []
```

```python
for file in os.listdir(path):
    d = os.path.join(path, file)
    if os.path.isdir(d):
        print(d)
        dirpath = d
        filepath = d + '/p' 
        try :
            for file_name in os.listdir(filepath):
                html_doc = open(filepath + "/" + file_name, 'rt', encoding='UTF8')
                soup = bs(html_doc, 'html.parser')
                html_doc.close() 
                try:
                    product = soup.find('div', attrs={ 'id' : "single-product" })
                    
                    p_file_name_list.append(file_name) # 여기다 써줘야 login 페이지나 깨진 페이지는 안 들어감
                    
                    try :
                        product_name_list.append(str(soup.title.string))
                    except :
                        product_name_list.append("")
                    try:
                        vendor_name = str(product.find('a', attrs={ 'class' : 'gen-user-link'}).get('href'))
                        vendor_name = vendor_name.split("/")[-1]
                        vendor_name = vendor_name.split("#")[0]
                        vendor_name_list.append(vendor_name)
                    except:
                        vendor_name_list.append("")

                    try:
                        src_file = d + str(product.find('img').get("src"))
                        img_src_list.append(src_file)
                    except:
                        img_src_list.append("")

                    try:
                        value = str(product.find('div', attrs={ 'class' : 'product-page-price'}).text)
                        value = value.split(" ")[0]
                        value = float(value)
                        value_list.append(value)
                    except:
                        value_list.append("") 

                    try:
                        ships = str(product.find('div', attrs={ 'class' : 'product-page-ships'}).text)
                        try:
                            from_value = ships.split("From: ")[1]
                        except:
                            from_value = ""
                        try:
                            from_value = from_value.split("To: ")[0]
                        except:
                            # print("only FROM exists")
                            pass
                        try:
                            to_value = ships.split("To: ")[1]
                        except:
                            to_value = ""
                        
                        from_value = from_value.replace(" ", "")
                        from_value = from_value.replace("\n", "")
                        to_value = to_value.replace(" ", "")
                        to_value = to_value.replace("\n", "")

                        from_list.append(from_value)
                        to_list.append(to_value)
                    except:
                        # 이건 안 해줘도 되나..?
                        from_list.append("") 
                        to_list.append("")
                        
                    try:
                        product.find('div', attrs={ 'class' : 'embedded-feedback-list' }).decompose()
                        context_list.append(str(product.text))
                    except:
                        context_list.append("")
                        
                    try:
                        category1_list.append(str(soup.findAll('div', attrs={ 'class' : 'topnav-element'})[0].text))
                    except:
                        category1_list.append("")
                    try:
                        category2_list.append(str(soup.findAll('div', attrs={ 'class' : 'topnav-element'})[1].text))
                    except:
                        category2_list.append("")
                    try:
                        category3_list.append(str(soup.findAll('div', attrs={ 'class' : 'topnav-element'})[2].text))
                    except:
                        category3_list.append("")

                except:
                    pass
        
        except FileNotFoundError as e: 
                print(e)
```

```python
print(len(p_file_name_list))
print(len(product_name_list))
print(len(vendor_name_list))
print(len(img_src_list))
print(len(value_list))
print(len(from_list))
print(len(to_list))
print(len(context_list))
print(len(category1_list))
print(len(category2_list))
print(len(category3_list))
```

```python
# file_name : .html 붙어있는 애들은 제거해주기

search = ".html"

for i, p_file_name in enumerate(p_file_name_list):
    if search in p_file_name:
        p_file_name_list[i] = p_file_name.strip(search)
```

## Dataframe 만들기

```python
df1 = pd.DataFrame({'File Name': p_file_name_list, 'Product Name': product_name_list, 'Vendor': vendor_name_list, 'IMG src': img_src_list, 'Value': value_list, 'From': from_list, 'To': to_list , 'Context': context_list, 'Category 1': category1_list, 'Category 2': category2_list , 'Category 3': category3_list})
# csv로 저장해두면 (백업) 나중에 그냥 이거 불러와도 됨
df1.to_csv('./agora/product.csv', index=False)
df1.tail(20)
```

# Vendor

```python
v_file_name_list = []
v_vendor_name_list = []
PGP_key_list = []
total_deals_list = []
rating_list = []
verified_list = []
success_deals_list = []
regist_date_list = []
```

```python
for file in os.listdir(path):
    d = os.path.join(path, file)
    
    if os.path.isdir(d):
        print(d)
        dirpath = d
        filepath = d + '/vendor' 
        try :
            for file_name in os.listdir(filepath):
                html_doc = open(filepath + "/" + file_name, 'rt', encoding='UTF8')
                soup = bs(html_doc, 'html.parser')
                try:
                    vendor = soup.find('div', attrs={ 'id' : "middlestuff" })
                    
                    v_file_name_list.append(file_name) # 여기다 써줘야 깨진 페이지는 안 들어감
                    
                    try :
                        v_vendor_name_list.append(str(vendor.find('strong').text))
                    except :
                        v_vendor_name_list.append("")
                        
                    try: 
                        pgp = str(vendor.find('span', attrs={ 'class' : 'pgptoken' }).text)
                        PGP_key_list.append(pgp.split('-----')[2])
                    except:
                        try:
                            pgp = str(vendor.find('textarea', attrs={ 'class' : 'pgpkeytoken' }).text)
                            PGP_key_list.append(pgp.split('-----')[2]) 
                        except:
                            PGP_key_list.append("")
                    
                    try:
                        total_deals = str(vendor.find('span', attrs={ 'class' : 'gen-user-ratings' }).text)
                        try:
                            N = re.findall(r"\d+", total_deals.split(',')[1])
                            try:
                                total_deals.split("~")
                                total_deals_list.append((int(N[-2]) + int(N[-1])) / 2)   # 1~2 deals 로 표기될 땐 평균값 추출
                            except:
                                total_deals_list.append(int(N[-1]))
                        except:
                            N = re.findall(r"\d+", total_deals)
                            total_deals_list.append(int(N[-1]))
                    except:
                        total_deals_list.append("")
                    
                    try:
                        rating = str(vendor.find('span', attrs={ 'class' : 'gen-user-ratings' }).text)
                        N = re.findall(r"\d+", rating.split(',')[0])
                        if len(N) == 3:
                            tmp = N[0] + "." + N[1]
                            rating_list.append( float(tmp) / float(N[-1]) )
                        else:
                            rating_list.append( int(N[-2]) / int(N[-2]) )
                    except:
                        rating_list.append("")

                    try:
                        a = vendor.find('div', attrs={ 'class' : 'vendor-verification' })
                        a.find('span', attrs={ 'class' : 'extra-info' }).decompose()
                        tmp = str(a.text)
                        verified_list.append(tmp.split(": ")[1])
                    except:
                        verified_list.append("")
                    
                    try:
                        success_deals = str(vendor.find('span', attrs={ 'class' : 'extra-info' }).text)
                        success_deals_list.append(success_deals.split(" ")[0])
                    except:
                        success_deals_list.append("")
                    
                    try:
                        date = str(vendor.find('div', attrs={ 'class' : 'vendorbio-stats-online'}).text)
                        numbers = re.findall(r"\d+", date.split("Last seen")[0]) 
                        try : 
                            day = int(numbers[-1])
                        except:
                            day = 0
                        try:
                            month = int(numbers[-2])
                        except: 
                            month = 0
                        try:
                            year = int(numbers[-3])
                        except:
                            year = 0
                            
                        folder = file
                        folder_date = datetime.strptime(folder, "%Y-%m-%d")
                        regist_date = folder_date - relativedelta(years=year, months=month, days=day)
                            
                        #date_list.append(regist_date.strftime("%Y-%m-%d"))   -> string 타입으로 저장
                        regist_date_list.append(regist_date.date())  # date 타입으로 저장
                    except:
                        regist_date_list.append("")
                except:
                    pass
                
        except FileNotFoundError as e: 
                print(e)
```

```python
print(len(v_file_name_list))
print(len(v_vendor_name_list))
print(len(PGP_key_list))
print(len(total_deals_list))
print(len(success_deals_list))
print(len(verified_list))
print(len(rating_list))
print(len(regist_date_list))
```

```python
# file_name : .html 붙어있는 애들은 제거해주기

search = ".html"

for i, v_file_name in enumerate(v_file_name_list):
    if search in v_file_name:
        v_file_name_list[i] = v_file_name.strip(search)
```

## Dataframe 만들기

```python
df2 = pd.DataFrame({'File Name': v_file_name_list, 'Vendor Name': v_vendor_name_list, 'PGP key': PGP_key_list, 'Rating': rating_list, 'Total Deals': total_deals_list, 'Registration Date':  regist_date_list, 'Verified': verified_list, 'Successful Deals': success_deals_list })
df2.to_csv('./agora/vendor.csv', index=False)
df2.head(20)
df2.tail(20)
```

# Feedback

```python
f_vendor_name_list = []
product_file_list = []
f_value_list = []
date_list = []
amount_list = []
f_context_list = []
```

```python
import time
from dateutil.relativedelta import relativedelta
from datetime import datetime

for file in os.listdir(path):
    d = os.path.join(path, file)
    
    if os.path.isdir(d):
        print(d)
        filepath = d + '/userfeedbacks' 
        
        try :
            for dir_name in os.listdir(filepath):
                f_path = filepath + "/" + dir_name
                
                for file_name in os.listdir(f_path) :
                    full_name = f_path + "/" + file_name
                    html_doc = open(full_name, 'rt', encoding='UTF8')
                    soup = bs(html_doc, 'html.parser')
                    
                    try:
                        feedback_list = soup.find('div', attrs={ 'class' : 'embedded-feedback-list' }).find('table').findAll('tr')
                    except:
                        pass
                    
                    for feedback in feedback_list: 
                        f_vendor_name_list.append(dir_name) #여기서는 오류가 날 수 없으니까 굳이 다로 try-except문 안에 넣지 않아도 x
                        
                        try:
                            p_path = d + str(feedback.find('a').attrs['href'])
                            product_file_list.append(p_path)
                        except:
                            product_file_list.append("")
                        
                        try:
                            f_value = str(feedback.find('strong').text)
                            n = f_value.split('/')
                            f_value_list.append(float(n[0]) / float(n[1]))
                        except:
                            f_value_list.append("")
                            
                        try:
                            date = str(feedback.findAll('td')[3].text)
                            numbers = re.findall(r"\d+", date)
                            try : 
                                day = int(numbers[-1])
                            except:
                                day = 0
                            try:
                                month = int(numbers[-2])
                            except: 
                                month = 0
                            try:
                                year = int(numbers[-3])
                            except:
                                year = 0
                            
                            folder = file
                            folder_date = datetime.strptime(folder, "%Y-%m-%d")
                            
                            feedback_date = folder_date - relativedelta(years=year, months=month, days=day)
                            
                            #date_list.append(feedback_date.strftime("%Y-%m-%d"))   -> string 타입으로 저장
                            date_list.append(feedback_date.date())  # date 타입으로 저장
                        except:
                            date_list.append("")
                        
                        try:
                            amount = str(feedback.find('span', attrs={ 'class' : "gen-user-ratings" }).text)
                            N = re.findall(r"\d+", amount)  
                            try:
                                amount.split("~")
                                amount_list.append( (int(N[-2]) + int(N[-1])) / 2 )   # 1~2 deals 로 표기될 땐 평균값 추출
                            except:
                                amount_list.append(int(N[-1]))
                        except:
                            amount_list.append("")
                        
                        try:
                            f_context_list.append(str(feedback.findAll('td')[1].text))
                        except: 
                            f_context_list.append("")
                            
        except FileNotFoundError as e: 
                print(e)
```

```python
print(len(f_vendor_name_list))
print(len(product_file_list))
print(len(f_value_list))
print(len(date_list))
print(len(amount_list))
print(len(f_context_list))
```

```python
df3 = pd.DataFrame({'Vendor': f_vendor_name_list, 'Product File': product_file_list, 'Value': f_value_list, 'Date': date_list, 'Amount': amount_list, 'Context': f_context_list})
df3.to_csv('./agora/feedback.csv', index=False)
df3.head(10)
df3.tail(20)
```

# 중복 제거

```python
product = df1
vendor = df2
feedback = df3
```

```python
# 모든 feature 다 같으면 제거 : 마지막 꺼를 남겨둬야 최신 파일이 뭔지 판별 가능 (수정된 내용 중복 제거할 때 필요)
product = product.drop_duplicates(keep = 'last', ignore_index = True)                  #file name이 (아마) 인코딩된 값이라 중복이 없음
vendor = vendor.drop_duplicates(keep = 'last', ignore_index = True)
feedback = feedback.drop_duplicates(keep = 'last', ignore_index = True)
```

```python
product = product[product['Product Name'] != "Login"]
product = product.reset_index()
```

```python
# product는 이미지 소스가 절대경로라 같은 파일이어도 이미지 소스는 다르게 저장되어 있음 이거 제외하고 다 같으면 같은 파일로 보자 
# 똑같은 파일이 여러 카테고리에 다 게시되어 있으면 어차피 그 카테고리는 의미가 없으니까 Category 1, 2, 3도 비교군에서 빼줬음

product = product.drop_duplicates(['Product Name', 'Vendor', 'Value', 'From', 'To', 'Context'], keep = 'last', ignore_index = True)
```

```python
# 전부 다 같은데 value만 다른 건 가격이 걍 오른 경우일 수 있음 => 최신 것만 남기자
# Context 까지 포함해서 중복을 찾으면 없다고 떠서 Context도 제거함 (Context 안에 가격 정보가 포함됨)

product = product.drop_duplicates(['Product Name', 'Vendor', 'From', 'To', 'Category 1', 'Category 2', 'Category 3'], keep = 'last', ignore_index = True)
```
