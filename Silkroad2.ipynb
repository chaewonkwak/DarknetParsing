```python
for file in os.listdir(path):
    d = os.path.join(path, file)
    if os.path.isdir(d):
        print(d)
        dirpath = d
        filepath = d + '/items' 
        try :
            for file_name in os.listdir(filepath):
                html_doc = open(filepath + "/" + file_name, 'rt', encoding='UTF8')
                try:
                    soup = bs(html_doc, 'html.parser')
                    html_doc.close() 
                    try:
                        product = soup.find('div', attrs={ 'class' : "body" })
                        p_file_name_list.append(file_name)

                        try :
                            product_name_list.append(str(product.find('h2').text))
                        except :
                            product_name_list.append("")
                        try:
                            vendor_name = str(product.find('h3').find('a').text) #h3 태그 안의 a 태그 (h3 없으면 vendor 정보도 x)
                            vendor_name_list.append(vendor_name)
                        except:
                            vendor_name_list.append("")

                        try:
                            src_file = d + "/images/" + str(product.find('div', attrs={'class':'item_image'}).get('id'))
                            img_src_list.append(src_file)
                        except:
                            img_src_list.append("")

                        try:
                            price = str(product.find('div', attrs={'class':'price_big'}).text)
                            numbers = re.findall(r"\d+", price)
                            price_list.append(float(numbers[0] + "." + numbers[1]))
                        except:
                            price_list.append("") 

                        try:
                            description_list.append(str(product.find('div', attrs={'class':'container container_large'}).text).strip())
                        except:
                            description_list.append("")

                        # -----------category 최대 6개 1->6 순으로 상위->하위 카테고리---------------
                        # category는 product body 밖에 있으므로 product.find가 아니라 soup.find 해줘야 함
                        try:
                            full_categories = str(soup.find('div', attrs={'class':'categories'}).findAll('a')[0].get('href'))
                            tmp = full_categories.split("/")[2]
                            categories = tmp.split("-")                     # 배열로 담김. 

                            # 카테고리를 한번에 try-except 처리해주면 없는 값에 정확히 "" 넣기 복잡함

                            try:
                                category1_list.append(categories[0])
                            except:
                                category1_list.append("")
                            try:
                                category2_list.append(categories[1])
                            except:                                         # IndexError만으로 핸들링 가능할 거 같긴 한데..
                                category2_list.append("")
                            try:
                                category3_list.append(categories[2])
                            except:
                                category3_list.append("")
                            try:
                                category4_list.append(categories[3])
                            except:
                                category4_list.append("")
                            try:
                                category5_list.append(categories[4])
                            except:
                                category5_list.append("")
                            try:
                                category6_list.append(categories[5])
                            except:
                                category6_list.append("")
                        except:
                            category1_list.append("")
                            category2_list.append("")
                            category3_list.append("")
                            category4_list.append("")
                            category5_list.append("")
                            category6_list.append("")

                        #----------------------------------------------------------------------------

                        try:
                            ships = str(product.findAll('p')[1].text)
                            try:
                                from_value = ships.split("ships from: ")[1]
                            except:
                                from_value = ""
                            try:
                                from_value = from_value.split("ships to: ")[0]
                            except:
                                # print("only FROM exists")
                                pass
                            try:
                                to_value = ships.split("ships to: ")[1]
                            except:
                                to_value = ""

                            from_list.append(from_value.strip())
                            to_list.append(to_value.strip())
                        except:
                            from_list.append("") 
                            to_list.append("")
                    except:
                        pass
                except:
                    pass
        
        except FileNotFoundError as e: 
                print(e)
```
