## Parsing

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from pandas import DataFrame
import os
#from datetime import datetime
```

```python
path = "D:/blackbankmarket/"
```

```python
vendor_list = []
market_list = []
link_list = []
confirmed_list = []
verified_list = []
reviewed_on_list = []
```

```python
for file in os.listdir(path):
    d = os.path.join(path, file)
    if os.path.isdir(d):
        print(d)
        filepath = d + '/vendor' 
        try :
            for file_name in os.listdir(filepath):
                if file_name[-9:] == "portfolio":
                    html_doc = open(filepath + "/" + file_name, 'rt', encoding='UTF8')
                    soup = bs(html_doc, 'html.parser')
                    html_doc.close() 
                    
                    market = soup.findAll('div', attrs={'class' : 'inline dark-blue heavy'})
                    link = soup.findAll('textarea', attrs={'class': 'tiny'})
                    button = soup.findAll('button', {'class':['highlight-dark round white small', 'panel-green-selected round white small']})
                    date = soup.findAll('div', attrs={'class': 'inline', 'style' : 'width: 175px;'})
                    
                    try :
                        for i in range(len(market)):
                            try:
                                vendor_list.append(str(soup.find('span', attrs={'class' : 't1'}).text))
                            except:
                                vendor_list.append("")
                                
                            try:
                                market_list.append(str(market[i].text.strip()))
                            except:
                                market_list.append("")
                                
                            try:
                                link_list.append(str(link[i].text))
                            except:
                                link_list.append("")
                                
                            try:
                                a = str(button[i*2].text)
                                b = str(button[i*2+1].text)
                                try:
                                    if(a == "Confirmed"):
                                        confirmed_list.append('Y')
                                    else:
                                        confirmed_list.append('N')
                                except:
                                    confirmed_list.append("")
                                try:
                                    if(b == "Verified"):
                                        verified_list.append('Y')
                                    else:
                                        verified_list.append('N')
                                except:
                                    verified_list.append("")
                            except:
                                confirmed_list.append("")
                                verified_list.append("")

                            try:
                                reviewed_on_list.append(str(date[3+i*2].text.strip()))
                            except:
                                reviewed_on_list.append("")
                    except : 
                        pass
        
        except FileNotFoundError as e: 
                print(e)
```

## Dataframe 만들기

```python
for i in range(len(market_list)):
    if market_list[i] =='Agora Market':
        market_list[i] = 'Agora'

if 'Agora Market' in market_list:
    print('Not Yet')
else:
    print('done')
```

```python
df = pd.DataFrame({'Vendor': vendor_list, 'Market': market_list, 'Link': link_list, 'Confirmed': confirmed_list, 'Verified': verified_list, 'Reviewed on': reviewed_on_list })
df.to_csv('./blackbankmarket/blackbankmarket.csv', index=False)
df.head(20)
df.tail(20)
```
